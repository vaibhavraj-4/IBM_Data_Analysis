```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("education.csv")
data.head()
```

```python
#checking for missing value and inconsistent data

#checking for missing value and inconsistent data
missing_values = data.isnull().sum()
inconsistent_data = data.apply(lambda x: sum(x == 'NR'))

#stabdardize column names
data.columns = [c.strip().replace(' ', '_').replace('__', '_').lower() for c in data.columns] # Changed col to c

# Convert data types if necessary (e.g., numeric columns)
numeric_columns = [
    'literacy_rate', 'enrollment_primaryboys', 'enrollment_primarygirls',
    'enrollment_primary_total', 'enrollment_secondary_boys', 'enrollment_secondar_girls',
    'enrollment_secondary_total', 'dropout_primary_boys', 'dropout_primary_girls',
    'dropout_primary_total', 'dropout_secondary_boysdropout_secondary_boy',
    'dropout_secondary_girls', 'dropout_secondary_total', 'schools_with_girsl_toilet',
    'schools_with_boys_toilet'
]

for col in numeric_columns:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Handle missing or inconsistent data
data = data.replace('NR', pd.NA)
data = data.replace(0, pd.NA)
data = data.dropna()

# Display the cleaned data
data

```

```python
# Save the cleaned data to a new CSV file
data.to_csv("new_education_analysis.csv")
```

```python
# Set the style for seaborn
sns.set(style="whitegrid")
```

```python
# Scatter plot: Literacy Rate vs Enrollment Primary Total
plt.figure(figsize=(10, 6))
sns.scatterplot(x='literacy_rate', y='enrollment_primary_total', data=data)
plt.title('Literacy Rate vs Enrollment Primary Total')
plt.xlabel('Literacy Rate')
plt.ylabel('Enrollment Primary Total')
plt.show()

# Scatter plot: Literacy Rate vs Dropout Primary Total
plt.figure(figsize=(10, 6))
sns.scatterplot(x='literacy_rate', y='dropout_primary_total', data=data)
plt.title('Literacy Rate vs Dropout Primary Total')
plt.xlabel('Literacy Rate')
plt.ylabel('Dropout Primary Total')
plt.show()

# Scatter plot: Enrollment Primary Total vs Dropout Primary Total
plt.figure(figsize=(10, 6))
sns.scatterplot(x='enrollment_primary_total', y='dropout_primary_total', data=data)
plt.title('Enrollment Primary Total vs Dropout Primary Total')
plt.xlabel('Enrollment Primary Total')
plt.ylabel('Dropout Primary Total')
plt.show()
```

```python
# Literacy rate distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['literacy_rate'], kde=True, bins=20)
plt.title('Distribution of Literacy Rate')
plt.xlabel('Literacy Rate')
plt.ylabel('Frequency')
plt.show()
```

```python
# Enrollment rates by gender
plt.figure(figsize=(12, 6))
sns.barplot(x='location', y='enrollment_primaryboys', data=data, label='Boys', color='b', ci=None)
sns.barplot(x='location', y='enrollment_primarygirls', data=data, label='Girls', color='r', ci=None)
plt.title('Primary Enrollment Rates by Gender')
plt.xlabel('Location')
plt.ylabel('Enrollment Rate')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
# Dropout rates by gender
plt.figure(figsize=(12, 6))
sns.barplot(x='location', y='dropout_primary_boys', data=data, label='Boys', color='b', ci=None)
sns.barplot(x='location', y='dropout_primary_girls', data=data, label='Girls', color='r', ci=None)
plt.title('Primary Dropout Rates by Gender')
plt.xlabel('Location')
plt.ylabel('Dropout Rate')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
# Availability of school facilities
plt.figure(figsize=(12, 6))
sns.barplot(x='location', y='schools_with_girsl_toilet', data=data, label='Girls Toilet', color='g', ci=None)
sns.barplot(x='location', y='schools_with_boys_toilet', data=data, label='Boys Toilet', color='y', ci=None)
plt.title('Availability of School Toilets')
plt.xlabel('Location')
plt.ylabel('Percentage of Schools')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
